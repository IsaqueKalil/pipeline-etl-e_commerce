```python
import pandas as pd
df_usuarios = pd.read_csv('usuarios.csv')
df_compras = pd.read_csv('compras.csv')
```

```python
display(df_compras)
display(df_usuarios)
```

## LIMPANDO E TRATANDO A TABELA DE USUÁRIOS

```python
df_usuarios = df_usuarios.drop_duplicates()
df_usuarios = df_usuarios.drop(columns=['index'], errors='ignore')
df_usuarios['nome'] = df_usuarios['nome'].str.strip()
df_usuarios['cpf'] = df_usuarios['cpf'].str.replace('-', '').str.replace('.', '')
df_usuarios['cpf'] = pd.to_numeric(df_usuarios['cpf'])

```

```python
df_usuarios['data_cadastro'] = pd.to_datetime(df_usuarios['data_cadastro'])
df_usuarios.info()

df_usuarios['nome'] = df_usuarios['nome'].str.title()
display(df_usuarios)

```

## LIMPANDO E TRATANDO A TABELA DE COMPRAS

```python
df_compras['valor'] = df_compras['valor'].str.replace('R$', '')
df_compras['data_compra'] = df_compras['data_compra'].str.replace('/', '-')
df_compras['valor'] = pd.to_numeric(df_compras['valor'])
df_compras['data_compra'] = pd.to_datetime(df_compras['data_compra'])
df_compras['valor'] = df_compras['valor'].fillna(df_compras['valor'].mean())
display(df_compras)
df_compras.info()

```

```python
df_usuarios.to_csv('usuarios_limpos.csv', index=False)
df_compras.to_csv('compras_limpas.csv', index=False)
```
