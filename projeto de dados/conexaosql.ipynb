## CRIANDO O BANCO DE DADOS E SUAS TABELAS

```python
comando =f''' CREATE TABLE IF NOT EXISTS usuarios (          
          id_usuario int PRIMARY KEY AUTO_INCREMENT,
          nome varchar(100),
          email varchar(100),
          cpf bigint,
          data_cadastro date)'''
cursor.execute(comando)
conexao.commit()

cursor.close()
conexao.close()
```

```python
comando = f''' create table if not exists compras (
               id_compra int primary key auto_increment,
               id_usuario int,
               produto varchar(100),
               valor decimal(10,2),
               data_compra date,
               constraint fk_usuarios_compras foreign key (id_usuario) references usuarios (id_usuario)
               )
'''
cursor.execute(comando)
conexao.commit()


cursor.close()
conexao.close()
```

```python
comando = 'drop table compras'
cursor.execute(comando)
conexao.commit()
conexao.close()
cursor.close()
```
